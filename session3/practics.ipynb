# Data Cleaning

```python
def convert_date(date_string):
    from datetime import date
    tokens = date_string.split('/')
    return date(year=1911 + int(tokens[0]), month= int(tokens[1]), day=int(tokens[2]))
```

```python
convert_date('83/09/05')
```

```python
df['西元日期'] = df['交易日期'].apply(convert_date)
```

```python
df
```

```python
df_clean = df[['西元日期', '收盤價']]
```

```python
df_clean.set_index('西元日期')
```
